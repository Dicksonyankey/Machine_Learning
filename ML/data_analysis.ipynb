````python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, r2_score, mean_absolute_error, mean_squared_error
from sklearn.linear_model import LinearRegression 
````

````python
data = pd.read_csv("heart.csv")
data
````

````python
pref_data = data.loc[0:,["age","sex","cp","chol",'fbs']]
````

````python
pref_data
````

````python
pref_data.columns = ["Age","Sex","Cp_Value","Cholesterol","Fasting_Value"]
````

````python
pref_data
````

###  Splitting the dataset into Features and Target from our ML training set

````python
features = pref_data.loc[0:,["Cp_Value","Cholesterol","Fasting_Value"]]
target = pref_data.loc[0:, ["Sex"]]
````

````python
features
````

````python
# Checking the shape of the feature dataset fror the training
features.shape
````

````python
# Checking a simple statistical infomation on the feature dataset 

features.describe()
````

````python
# Checking the shape for the target dataset 
target.shape
````

````python
print(features.columns, target.columns)
````

````python
#  Checking a simple statistical infomation on the target dataset
target.describe()
````

````python
sns.displot(data=pref_data, x="Cholesterol")
````

````python
model = LinearRegression()
X_train, X_test, Y_train, Y_test = train_test_split(features, target, test_size=0.6, random_state=587)
X_train.shape, Y_train.shape
````

````python
model.fit(X_train, Y_train)
````

````python
Y_pred = model.predict(X_test)
````

````python
Y_pred
````

````python
r2_score(Y_test, Y_pred)
````

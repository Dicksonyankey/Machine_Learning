````python
# Import All Libraries need for the analysis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns

# This make all graphs presents in this notebook
%matplotlib inline 
````

````python
# Set figure size
plt.figure(figsize=(8, 5))
# Create the data
countries = ['Qatar', 'Luxembourg', 'Singapore', 'Brunei', 'Ireland', 'Norway', 'UAE', 'Kuwait']
populations = [2781682,604245, 5757499, 428963, 4818690, 5337962, 9630959, 4137312]
gdp_per_capita = [130475, 106705, 100345, 79530, 78785, 74356, 69382, 67000]

# scale GDP per capita income to shoot the bubbles in the graph
scaled_gdp_per_capita = np.divide(gdp_per_capita, 120)
colors = np.random.rand(8)

# Draw the scatter diagram
plt.scatter(countries, populations, s=scaled_gdp_per_capita, c=colors,cmap="Blues", edgecolors="grey", alpha=0.5)
# Add X Label on X-axis
plt.xlabel("Countries")
# Add Y Label on X-axis
plt.ylabel("Population")
# Add title to graph
plt.title("Bubble Chart")

# rotate x label for clear visualization
plt.xticks(rotation=45)
# Show the plot
plt.show()
````

# lm plots
The lm plot plots the scatter and fits the regression model on it. A scatter plot is the best way to understand the relationship between two variables. 

```python
sns.lmplot()
```

arguments that lm plot has to receive
```  
X - value -> string value
Y - value -> string value
data (ie) the dataset itslef
fit_reg = True by default but it can be set to False
```

````python
# Create DataFrame
df = pd.DataFrame({'x': [1, 3, 5, 7, 9, 11], 'y': [10, 25, 35, 33, 41, 59]})
# Create lmplot
sns.lmplot(x='x', y='y', data=df)
# Show figure
plt.show()
````

````python
# Create lmplot
sns.lmplot(x='x', y='y', data=df, fit_reg=False)
# Show figure
plt.show()
````

````python
# Getting all the names of the dataset.
print(sns.get_dataset_names())

# Loading the dataset
data = sns.load_dataset("tips")

# Plotting the barplot
# In the barplot you can pass in (estimator) to base your numerical axis base on the mean, median, or variance , std, cov
sns.barplot(data=data, x="sex", y="tip", estimator=np.mean)
plt.xticks(rotation=60)
plt.show()
````

````python
# Plotting a boxplot
sns.boxplot(data=data, x="total_bill", y="tip", hue="sex")
plt.show()
````

````python
# Plotting a boxplot
sns.boxplot(data=data, x="day", y="total_bill", hue="sex")
plt.show()
````

````python
# Plotting a violinplot 
# You specific the split or not
sns.violinplot(data=data, x="day", y="total_bill", hue="sex", split=True)
plt.show()
````

````python
# Plotting a countplot 
sns.countplot(x="sex", data=data)
plt.show()
````

````python
# Plotting a pairplot 
# The X and Y Label can be overrided
# And other matplotlib functions can be called as well
sns.pairplot(data, hue="sex", palette="Blues")
plt.show()
````

````python
# Plotting a stripplot

# To change to size of the plot 
plt.figure(figsize=(10,5))

# strippolt
# you can pass jitter=True -> This will spread the data out a little bit.
# You can as well pass in hue="sex"
# You can also passs in dodge=True -> This will separate the Male and Female data

sns.stripplot(x="day", y="total_bill", data=data, jitter=True, hue="sex", dodge=True)
plt.show()
````

````python
# Plotting a swarm plot with Violinplot

sns.violinplot(data=data, x="day", y="total_bill")
sns.swarmplot(data=data, x="day", y="total_bill" color="white")
plt.show()
````

````python
# Plotting a swarm plot
sns.swarmplot(data=data, x="day", y="total_bill" color="white")
plt.show()

````

````python
# Plotting a Platteplot
plt.figure(figsize=(10,6))
sns.set_style("white") # you can pass in paper to see what happens
sns.set_context("talk")
sns.stripplot(x="day", y="total_bill", data=data,jitter=True, hue="sex", dodge=True, palette="afmhot")
plt.show()
````

## HEATMAPS

````python
# Using the car_crashes dataset from seaborn 

crash_df = sns.load_dataset("car_crashes")

# Getting the correlation from the crashes dataset 

crash_corr = crash_df.corr()
crash_corr
````

````python
# Plotting a heatmap from the crash_corr
plt.figure(figsize=(10,5))
sns.set_context("paper", font_scale=1.5)
# plotting
sns.heatmap(crash_corr, annot=True, cmap="Blues")
plt.show()
````
